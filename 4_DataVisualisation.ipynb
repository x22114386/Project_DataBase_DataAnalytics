```python
import matplotlib.pyplot as plt
import pandas as pd

energy_data = pd.read_csv("C:\\Users\\karth\\PERSONAL\\Data_Analytics\\DAP\\PROJECT_CA\\CleanedEnergyEfficiency.csv")

# Plot the result as a bar plot
plt.bar(energy_data['Pre-Retrofit Home Heating Fuel Type'],energy_data['Estimated Annual kWh Savings'])

plt.xticks(rotation=90)

# Add labels and a title
plt.ylabel('Savings (kWh)')
plt.xlabel('Fuel Type')
plt.title('Estimated Annual Energy Savings')
plt.ylim([0,13000])
plt.show()
```

Pie Chart

```python
import matplotlib.pyplot as plt


# Getting the frequency of each word in the column
value1 = energy_data['Pre-Retrofit Home Heating Fuel Type'].loc[energy_data['Pre-Retrofit Home Heating Fuel Type'] == 'Kerosene'].count()
value2 = energy_data['Pre-Retrofit Home Heating Fuel Type'].loc[energy_data['Pre-Retrofit Home Heating Fuel Type'] == 'Electric'].count()
value3 = energy_data['Pre-Retrofit Home Heating Fuel Type'].loc[energy_data['Pre-Retrofit Home Heating Fuel Type'] == 'Wood'].count()
value4 = energy_data['Pre-Retrofit Home Heating Fuel Type'].loc[energy_data['Pre-Retrofit Home Heating Fuel Type'] == 'Propane'].count()
value5 = energy_data['Pre-Retrofit Home Heating Fuel Type'].loc[energy_data['Pre-Retrofit Home Heating Fuel Type'] == 'Natural Gas'].count()
print(value5)

# Create a list of values and labels
values = [value1, value2, value3, value4, value5]
labels = ['Kerosene', 'Electric', 'Wood', 'Propane', 'Natural gas']

# Creating a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')

plt.title('Heating Fuel Type used in homes')

plt.show()
```

Area Chart

```python
areaData = pd.read_csv(
    "C:\\Users\\karth\\PERSONAL\\Data_Analytics\\DAP\\PROJECT_CA\\CleanedEnergyEfficiency - Copy.csv", 
    usecols=["Project Completion Date","Estimated Annual MMBtu Savings"],
)
areaData.replace(" ", np.nan, inplace=True)
# convert the date column to datetime format
areaData['Project Completion Date'] = pd.to_datetime(areaData['Project Completion Date'], format='%Y/%m/%d')

filtered_data = areaData.loc["2020-01-01":"2022-12-30"]

fig = px.area(
    filtered_data, 
    x=filtered_data.index, 
    y="Estimated Annual MMBtu Savings"
).update_layout(
    xaxis_title="Date",
    yaxis_title="Energy Savings"
)
fig.show()
```
