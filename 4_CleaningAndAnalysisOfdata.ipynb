State wise which is the Gas that is emitted maximum and what is the amount that is emitted

```python
import pandas as pd

# Group the dataset by state and pollutant_gas, and calculate the sum of pollutant_max for each group
grouped = df.groupby(['state', 'pollutant_id'])['pollutant_max'].sum().reset_index()

# Find the maximum pollutant_max value for each state
max_pollutant = grouped.groupby('state')['pollutant_max'].max().reset_index()

# Merge the maximum pollutant_max value with the original dataset to get the corresponding pollutant_gas value
result = pd.merge(max_pollutant, grouped, on=['state', 'pollutant_max'], how='left')

# Print the result
print(result[['state', 'pollutant_id', 'pollutant_max']])
```

Plotting a bar graph on States that has Maximum gasses emitted

```python
import matplotlib.pyplot as plt

# Plot the result as a bar plot
plt.bar(result['state'], result['pollutant_max'])

# Add labels and a title
plt.xlabel('State')
plt.ylabel('Maximum Pollutant Value')
plt.title('Maximum Pollutant Values by State')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
import pandas as pd


# Group the dataset by state and pollutant_gas, and calculate the sum of pollutant_min for each group
grouped = df.groupby(['state', 'pollutant_id'])['pollutant_min'].sum().reset_index()

# Find the minimum pollutant_min value for each state
min_pollutant = grouped.groupby('state')['pollutant_min'].min().reset_index()

# Merge the minimum pollutant_min value with the original dataset to get the corresponding pollutant_gas value
result = pd.merge(min_pollutant, grouped, on=['state', 'pollutant_min'], how='left')

# Print the result
print(result[['state', 'pollutant_id', 'pollutant_min']])
```

Produce a HeatMap

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm

# Group data by state and city and calculate average pollutant value
df_agg = df.groupby(['state', 'city'])['pollutant_avg'].mean().reset_index()

# Drop rows with missing or invalid data in the pollutant_avg column
df_agg = df_agg.dropna(subset=['pollutant_avg'])

# Create base map using Matplotlib's scatter function
fig, ax = plt.subplots(figsize=(20, 18))
ax.scatter(df_agg['state'], df_agg['city'], zorder=1, alpha=0)

# Define color scale for pollutant values
norm = colors.Normalize(vmin=df_agg['pollutant_avg'].min(), vmax=df_agg['pollutant_avg'].max())
cmap = cm.get_cmap('YlOrRd')

# Plot heat map by assigning colors to pollutant values
for i, row in df_agg.iterrows():
    plt.scatter(row['state'], row['city'], c=row['pollutant_avg'], cmap=cmap, norm=norm, edgecolor='none', zorder=2)

# Add colorbar legend and axis labels
cbar = plt.colorbar()
cbar.set_label('Pollutant Value')
plt.xlabel('State', fontsize=12)
plt.ylabel('City', fontsize=12)

# Rotate x-axis labels by 90 degrees and adjust spacing between subplots
plt.xticks(rotation=90, ha='center')
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Show the plot
plt.show()
```

State that is Most polluted

```python
# Load pollution data
pollution_data = df

# Calculate the average pollutant value for each state for each pollutant
state_pollution = pd.DataFrame(columns=['state', 'pollutant_id', 'pollutant_avg'])
for pollutant_gas in pollution_data['pollutant_id'].unique():
    pollutant_data = pollution_data[pollution_data['pollutant_id'] == pollutant_gas]
    state_pollution_by_pollutant = pollutant_data.groupby('state')['pollutant_avg'].mean().reset_index()
    state_pollution_by_pollutant['pollutant_id'] = pollutant_gas
    state_pollution = pd.concat([state_pollution, state_pollution_by_pollutant])

# Calculate the overall average pollutant value for each state
state_pollution = state_pollution.groupby('state')['pollutant_avg'].mean().reset_index()

# Sort states by overall pollutant value in descending order and select the first row (which will be the most polluted state)
most_polluted_state = state_pollution.sort_values(by='pollutant_avg', ascending=False).iloc[0]['state']

print(most_polluted_state)
```

State with More emission recieving station

```python
# Load pollution data
pollution_data = df

# Count the number of pollutant stations for each state
state_station_counts = pollution_data.groupby(['state', 'station'])['pollutant_id'].count().reset_index()
state_station_counts = state_station_counts.groupby('state')['station'].count().reset_index()
state_station_counts = state_station_counts.rename(columns={'station': 'station_count'})

# Find the state with the most pollutant stations
state_with_most_stations = state_station_counts.sort_values(by='station_count', ascending=False).iloc[0]['state']

print(state_with_most_stations)
```

State with Most type of gas emitted

```python
# Load pollution data
pollution_data = df

# Count the number of pollutant gases for each state
state_gas_counts = pollution_data.groupby(['state'])['pollutant_id'].nunique().reset_index()
state_gas_counts = state_gas_counts.rename(columns={'pollutant_id': 'gas_count'})

# Find the state with the most types of pollutant gas
state_with_most_gases = state_gas_counts.sort_values(by='gas_count', ascending=False).iloc[0]['state']

# Find the unique pollutant gases emitted in the state with the most types of gas
unique_gases = pollution_data[pollution_data['state'] == state_with_most_gases]['pollutant_id'].unique()

print("State with the most types of pollutant gas: {}".format(state_with_most_gases))
print("Unique pollutant gases emitted in this state: {}".format(list(unique_gases)))
```
