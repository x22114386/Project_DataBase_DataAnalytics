Merging four datasets

```python
import pandas as pd

#Defining four datasets csv file
file_names = ["CleanVehData.csv", "CleanedCompetitions.csv", "CleanedEnergyEfficiency.csv", "CleanedTreePlantations.csv"]

# Initializing an empty list to store the data frames
dfs = []

# Looping through the file names and loading the first 1000 rows of each CSV file into a data frame
for file_name in file_names:
    df = pd.read_csv(file_name, nrows=1000)
    dfs.append(df)

# Concatenating the data frames vertically (along the rows axis)
merged_df = pd.concat(dfs, axis=1)

#Replacing null values into NA
merged_df.fillna('NA', inplace=True)

# Write the merged dataset to a CSV file
merged_df.to_csv('MergedDataset.csv', index=False)

print('MergedDataset created.')
```

Visualising prominent data of each dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

#loading CSV file
df = pd.read_csv('MergedDataset.csv')

#Selecting specific values from each column to visualise
value1 = df['Electric Vehicle Type'].loc[df['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)'].count()
value2 = df['pollutant_id'].loc[df['pollutant_id'] == 'CO'].count()
value3 = df['Pre-Retrofit Home Heating Fuel Type'].loc[df['Pre-Retrofit Home Heating Fuel Type'] == 'Natural Gas'].count()
value4 = df['PlantType'].loc[df['PlantType'] == 'Tree'].count()


values = [value1, value2, value3, value4]
labels = ['Battery EV', 'Carbon Monoxide', 'Natural Gas', 'Trees planted']
colors = ['red', 'green', 'blue', 'yellow']

#=pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Prominent Data')
plt.show()
```
