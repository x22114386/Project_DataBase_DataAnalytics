```python
import csv


def getCsvRowChunks():
    chunkSize = 1000
    csvRowChunks = []

    with open('C:/Users/Manu/Downloads/StudiesNCI/MichealBrad_DB/projects/competitions.csv', 'r', encoding='utf-8') as csvfile:
        # Using DictReader becuase comma in some of the CSV column, normal read will split values based on comma
        csvreader = csv.DictReader(csvfile)
        output = list(csvreader)
        csvRowList = []
        for row in output:
            csvColumnList = []
            for key,val in row.items():
                if(key == "id"):
                    val = int(val)
                csvColumnList.append(val)
            csvRowList.append(csvColumnList)
            
            # chunking csv row lists
            if(len(csvRowList) >= chunkSize):
                csvRowChunks.append(csvRowList)
                csvRowList = []
            
        # Left over
        csvRowChunks.append(csvRowList)
        # print(csvRowChunks)
    return csvRowChunks
```
