```python
#HISTOGRAM

import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file into a pandas DataFrame
data = pd.read_csv('TreeplantCleaned.csv')

# Select column to create histogram from
column_name = 'Supervisor Districts'

# Create histogram
plt.hist(data[column_name])

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Histogram of ' + column_name)

# Display histogram
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#loading CSV file
df = pd.read_csv('TreeplantCleaned.csv')

#Selecting specific values from each column to visualise
value1 = df['qCaretaker'].loc[df['qCaretaker'] == 'Private'].count()
value2 = df['qCaretaker'].loc[df['qCaretaker'] == 'DPW'].count()
value3 = df['qCaretaker'].loc[df['qCaretaker'] == 'SFUSD'].count() 

values = [value1, value2, value3]
labels = ['PRIVATE', 'DPW', 'PRIVATE']
colors = ['red', 'green', 'blue']

#=pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('TREE CARETAKER')
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#loading CSV file
df = pd.read_csv('TreeplantCleaned.csv')

#Selecting specific values from each column to visualise
value1 = df['qLegalStatus'].loc[df['qLegalStatus'] == 'DPW Maintained'].count()
value2 = df['qLegalStatus'].loc[df['qLegalStatus'] == 'Undocumented'].count()
value3 = df['qLegalStatus'].loc[df['qLegalStatus'] == 'Permitted Site'].count() 

values = [value1, value2, value3]
labels = ['DPW Maintained', 'Undocumented', 'Permitted Site']
colors = ['yellow', 'green', 'red']

#=pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('LEGAL STSATUS OF TREES PLANTED')
plt.show() 
```

```python
#SCATTER PLOT

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('TreeplantCleaned.csv')

#Assigning x and y values from the dataframe
x = data['Supervisor Districts']
y = data['Police Districts']

#Scatterplot
plt.scatter(x, y)
plt.xlabel('Supervisor Districts')
plt.ylabel('Police Districts')
plt.title('Scatter Plot')
plt.show()
```
