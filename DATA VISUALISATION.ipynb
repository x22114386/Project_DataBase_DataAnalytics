```python
#HISTOGRAM

import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file into a pandas DataFrame
data = pd.read_csv('TreeplantCleaned.csv')

# Select column to create histogram from
column_name = 'Supervisor Districts'

# Create histogram
plt.hist(data[column_name])

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Histogram of ' + column_name)

# Display histogram
plt.show()
```

```python
#PIECHART

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('TreeplantCleaned.csv')

# Getting the frequency of each word in the column
counts = df['SiteOrder'].value_counts()

# Creating a pie chart
plt.pie(counts, labels=counts.index.tolist(), autopct='%1.1f%%')
plt.title('Pie Chart of site order for new trees to be planted')
plt.show()
```

```python
#SCATTER PLOT

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('TreeplantCleaned.csv')

#Assigning x and y values from the dataframe
x = data['Supervisor Districts']
y = data['Police Districts']

#Scatterplot
plt.scatter(x, y)
plt.xlabel('Supervisor Districts')
plt.ylabel('Police Districts')
plt.title('Scatter Plot')
plt.show()
```
